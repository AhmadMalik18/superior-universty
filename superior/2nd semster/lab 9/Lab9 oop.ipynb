```python
class Document:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
class Book(Document):
    def __init__(self, title, author, genre=None, pages=None):
        if genre and pages:
            self.title = title
            self.author = author
            self.genre = genre
            self.pages = pages
        else:
            self.title = title
            self.author = author
            self.genre = "Unknown"
            self.pages = 0

    def display_info(self):
        super().display_info()
        print(f"Genre: {self.genre}")
        print(f"Pages: {self.pages}")

    def set_details(self, title, author, genre=None, pages=None):
        if genre and pages:
            self.title = title
            self.author = author
            self.genre = genre
            self.pages = pages
        else:
            self.title = title
            self.author = author
            self.genre = "Unknown"
            self.pages = 0
class Article(Document):
    def __init__(self, title, author, journal=None, doi=None):
        if journal and doi:
            self.title = title
            self.author = author
            self.journal = journal
            self.doi = doi
        else:
            self.title = title
            self.author = author
            self.journal = "Unknown"
            self.doi = "Unknown"
    def display_info(self):
        super().display_info()
        print(f"Journal: {self.journal}")
        print(f"DOI: {self.doi}")
    def set_details(self, title, author, journal=None, doi=None):
        if journal and doi:
            self.title = title
            self.author = author
            self.journal = journal
            self.doi = doi
        else:
            self.title = title
            self.author = author
            self.journal = "Unknown"
            self.doi = "Unknown"
def save_to_csv(documents, filename="documents.csv"):
    with open(filename, mode="w", newline='') as file:
        writer = csv.writer(file)
        # Write header row
        writer.writerow(["Title", "Author", "Type", "Genre/Journal", "Pages/DOI"])
        
        for doc in documents:
            if isinstance(doc, Book):
                writer.writerow([doc.title, doc.author, "Book", doc.genre, doc.pages])
            elif isinstance(doc, Article):
                writer.writerow([doc.title, doc.author, "Article", doc.journal, doc.doi])
def load_from_csv(filename="documents.csv"):
    documents = []
    with open(filename, mode="r") as file:
        reader = csv.reader(file)
        next(reader) 
        
        for row in reader:
            title, author, doc_type, genre_or_journal, pages_or_doi = row
            if doc_type == "Book":
                genre = genre_or_journal
                try:
                    pages = int(pages_or_doi)
                except ValueError:
                    pages = 0 
                documents.append(Book(title, author, genre, pages))
            elif doc_type == "Article":
                # Handle loading Article
                journal = genre_or_journal
                doi = pages_or_doi
                documents.append(Article(title, author, journal, doi))
    
    return documents

if __name__ == "__main__":
    book1 = Book("MR.chips", "F. Scott", "Novel", 1180)
    book2 = Book("Freedom in the World", "Jamil Ahmad") 
    article1 = Article("Moth Smoke", "Mohsin Hamid")
    book1.display_info()
    print("\n")
    book2.display_info()
    print("\n")
    article1.display_info()
    print("\n")
    documents = [book1, book2, article1]
    save_to_csv(documents)
    loaded_documents = load_from_csv()
    for doc in loaded_documents:
        doc.display_info()
        print("\n")
```
