```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load image
image = cv2.imread('my_image.jpg')

# Convert BGR to RGB for Jupyter
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
# Save the image with a new name
cv2.imwrite('output_image.jpg', image)
```

```python
# Resize the image to 300x300
resized = cv2.resize(image, (300, 300))

# Rotate the image 45 degrees
center = (image.shape[1]//2, image.shape[0]//2)
rot_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, rot_matrix, (image.shape[1], image.shape[0]))

# Display the rotated image
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(rotated_rgb)
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Make a copy of the original image
img_copy = image.copy()

# Draw a blue line
cv2.line(img_copy, (50,50), (200,50), (255,0,0), 3)

# Draw a green rectangle
cv2.rectangle(img_copy, (50,70), (200,150), (0,255,0), 2)

# Draw a red circle
cv2.circle(img_copy, (150,300), 50, (0,0,255), -1)

# Add text in yellow
cv2.putText(img_copy, 'OpenCV', (50,400), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,0), 3)

# Display the result
plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian Blur
blurred = cv2.GaussianBlur(image, (7,7), 0)

# Perform Canny edge detection
edges = cv2.Canny(image, 100, 200)

# Display blurred image
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Blurred Image')
plt.show()

# Display edges
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title('Edge Detection')
plt.show()
```

```python
# Load the face cascade classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Convert image to grayscale for detection
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray_img, 1.1, 4)

# Draw rectangles around detected faces
img_faces = image.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(img_faces, (x,y), (x+w,y+h), (255,0,0), 2)

# Display the result
plt.imshow(cv2.cvtColor(img_faces, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
